```python
dataset_path = kagglehub.dataset_download('halaturkialotaibi/coffee-bean-sales-dataset')

print(dataset_path)
```

```python
os.listdir(dataset_path)
```

```python
sales_data = pd.read_csv(f'{dataset_path}/DatasetForCoffeeSales2.csv')
```

```python
len(sales_data)
```

```python
sales_data.head(20)
```

```python
sales_data.filter(['City' , 'Category' , 'Product' , 'Quantity']).head()
```

```python
sales_data.sort_values("Quantity" , axis=0, ascending=False , inplace = True ,  na_position = 'last')
display(sales_data.head())
```

```python
sales_data_groupby_city = sales_data.groupby('City')

result = sales_data_groupby_city['Final Sales'].agg('sum')
```
